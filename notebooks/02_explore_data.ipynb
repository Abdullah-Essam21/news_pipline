```python
# Arabic month → numeric mapping
month_map = {
    "يناير": "01",
    "فبراير": "02",
    "مارس": "03",
    "أبريل": "04",
    "مايو": "05",
    "يونيو": "06",
    "يوليو": "07",
    "أغسطس": "08",
    "سبتمبر": "09",
    "أكتوبر": "10",
    "نوفمبر": "11",
    "ديسمبر": "12",
}

# Step 1: clean + fix AM/PM safely
expr = (
    pl.col("publish_date")
    .str.replace(r"^.*،\s*", "")   # remove weekday الجزء الأول
    .str.replace(r"\sص$", " AM")   # replace only ending "ص"
    .str.replace(r"\sم$", " PM")   # replace only ending "م"
)

# Step 2: replace Arabic month names
for ar, num in month_map.items():
    expr = expr.str.replace(ar, num)

# Step 3: parse into datetime
df = df.with_columns(
    expr
    .str.strptime(pl.Datetime, "%d %m %Y %I:%M %p", strict=False)
    .alias("date_parsed")
)
```

```python
df['date_parsed', 'publish_date']
```

```python
ctx = pl.SQLContext(horoscope=df)
```

```python
result = ctx.execute("select url, content from horoscope where extract(month from date_parsed) = 3 and extract(day from date_parsed) = 21").collect()
result.unique().write_csv("../data/processed/march_21_horoscopes.csv", include_bom=True)
```
